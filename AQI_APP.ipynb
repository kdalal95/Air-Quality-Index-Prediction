```python
from tkinter import *
from PIL import ImageTk #to use jpg
```

```python
import pickle
```

```python
model= pickle.load(open('AQI_RFR_model.pkl', 'rb'))
```

```python
class AQI:
    def __init__(self,root):
        
        self.root= root
        root.iconbitmap(r'C:\Users\Krishna Dalal\Python Aegis\Projects\Air Quality Index Prediction\icon_aqi.ico')
        #Header
        self.root.title('Air Quality Index')
        
        #page size
        self.root.geometry('1199x600+150+100')
        self.root.resizable(False, False)
        
        #Image
        self.bg= ImageTk.PhotoImage(file='world.jpg')
        self.bg_image= Label(self.root, image= self.bg).place(x=0, y=0, relwidth=1, relheight=1)
        
        #frame
        Frame_AQI=Frame(self.root, bg= 'white')
        Frame_AQI.place(x=600, y=50, height=500, width= 500)
        
        
        #Title
        title= Label(Frame_AQI, text= 'AIR Quality Index', font=('Arial', 25, 'bold', 'underline'), fg='black', bg='white').place(x=120, y=18)
        #intro
        intro= Label(Frame_AQI, text= 'Fill in the details to calculate AQI :', font=('Arial', 15), fg='black', bg='white').place(x=40, y=75)
        
        
        #Variables and text box
        #Average temperature(T)
        T= Label(Frame_AQI, text= 'Average temperature', font=('Arial', 12), fg='black', bg='white').place(x=40, y=115)
        self.T_box= Entry(Frame_AQI, bg= 'lightgray')
        self.T_box.grid(row=1, column=1)
        self.T_box.place(x=235, y= 115, height= 20, width= 200)
        
        #maximum temperature(TM)
        TM= Label(Frame_AQI, text="Maximum temperature", font=("Arial",12), fg="black", bg="white").place(x=40, y=150)
        self.TM_box= Entry(Frame_AQI, bg= 'lightgray')
        self.TM_box.grid(row=1, column=1)
        self.TM_box.place(x=235, y= 150, height= 20, width= 200)
        
        #minimum temperature(tm)
        tm= Label(Frame_AQI,text="Minimum temperature",font=("Arial",12),fg="black",bg="white").place(x=40,y=185)
        self.tm_box= Entry(Frame_AQI, bg= 'lightgray')
        self.tm_box.grid(row=1, column=1)
        self.tm_box.place(x=235, y= 185, height= 20, width= 200)
        
        #Average relative humidity(H)
        H= Label(Frame_AQI,text="Average relative humidity",font=("Arial",12),fg="black",bg="white").place(x=40,y=220)
        self.H_box= Entry(Frame_AQI, bg= 'lightgray')
        self.H_box.grid(row=1, column=1)
        self.H_box.place(x=235, y= 220, height= 20, width= 200)
        
        #Total rainfall and / or snowmelt (mm) (PP)
        PP= Label(Frame_AQI,text="Total rainfall",font=("Arial",12),fg="black",bg="white").place(x=40,y=255)
        self.PP_box= Entry(Frame_AQI, bg= 'lightgray')
        self.PP_box.grid(row=1, column=1)
        self.PP_box.place(x=235, y= 255, height= 20, width= 200)
        
        #Average visibility (Km)(VV)
        VV= Label(Frame_AQI,text="Average visibility",font=("Arial",12),fg="black",bg="white").place(x=40,y=290)
        self.VV_box= Entry(Frame_AQI, bg= 'lightgray')
        self.VV_box.grid(row=1, column=1)
        self.VV_box.place(x=235, y= 290, height= 20, width= 200)
        
        #Average wind speed (Km/h)(V)
        V= Label(Frame_AQI,text="Average wind speed",font=("Arial",12),fg="black",bg="white").place(x=40,y=325)
        self.V_box= Entry(Frame_AQI, bg= 'lightgray')
        self.V_box.grid(row=1, column=1)
        self.V_box.place(x=235, y= 325, height= 20, width= 200)
        
        #Maximum sustained wind speed (Km/h)(VM)
        VM= Label(Frame_AQI,text="Max sustained windspeed",font=("Arial",12),fg="black",bg="white").place(x=40,y=360)
        self.VM_box= Entry(Frame_AQI, bg= 'lightgray')
        self.VM_box.grid(row=1, column=1)
        self.VM_box.place(x=235, y= 360, height= 20, width= 200)
        
        
        def data():
                T_val= float(self.T_box.get())
                TM_val= float(self.TM_box.get())
                tm_val= float(self.tm_box.get())
                H_val= float(self.H_box.get())
                PP_val= float(self.PP_box.get())
                VV_val= float(self.VV_box.get())
                V_val= float(self.V_box.get())
                VM_val= float(self.VM_box.get())
                values= [[T_val, TM_val, tm_val, H_val, PP_val, VV_val, V_val, VM_val]]
                pred_result= model.predict(values)
                pred_result= pred_result.round(2)
                
                if pred_result <=100:
                    prediction= Label(Frame_AQI, text= 'The Air Quality Index is {}'.format(pred_result), font=("Arial",22),fg='forest green', bg= 'white').place(x=52 , y=440)
                elif pred_result >100 and pred_result <=200:
                    prediction= Label(Frame_AQI, text= 'The Air Quality Index is {}'.format(pred_result), font=("Arial",22),fg='orange', bg= 'white').place(x=52 , y=440)
                else:
                    prediction= Label(Frame_AQI, text= 'The Air Quality Index is {}'.format(pred_result), font=("Arial",22),fg='red2', bg= 'white').place(x=52 , y=440)
        
        #Button
        button= Button(Frame_AQI, command= data, text= 'Predict', font=("Arial",10),fg="black",bg="grey").place(x=220,y=395)
        
root= Tk()
obj= AQI(root)
root.mainloop()
```
